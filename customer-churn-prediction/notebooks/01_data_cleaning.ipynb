```python
#import dependencies
#Load dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Path to the CSV file
csv_path = "../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv"

# Load into a DataFrame
df = pd.read_csv(csv_path)

# Quick peek
df.head()
```

```python
df.head()
df.info()
df.describe(include='all')
```

```python
df['TotalCharges'].dtype
```

```python
#convert TotalCharges column to be of numeric type so we can perform numerical analysis

# Replace empty strings or spaces with NaN
df['TotalCharges'] = df['TotalCharges'].replace(r'^\s*$', np.nan, regex=True)

# Convert to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.loc[df['TotalCharges'].isna(), 'TotalCharges'] = (
    df['MonthlyCharges'] * df['tenure']
)
```

```python
df.isna().sum()
```

```python
#Encode churn column
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
#Handle No internet service
cols_to_clean = ["OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport",
                 "StreamingTV","StreamingMovies"
]

for col in cols_to_clean:
    df[col] = df[col].replace('No internet service','No')
```

```python
#Save cleaned dataset
df.to_csv("../data/cleaned/cleaned_telco_customer_churn.csv", index=False)
```
